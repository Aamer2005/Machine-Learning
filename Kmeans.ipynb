```python
from math import sqrt
import random

class Kmeans:
    def __init__(self, k=3, iter=100):
        self.k = k
        self.iter = iter
        self.centroids = []

    def fit(self, X):
        # Initialize random centroids
        self.centroids = [X[random.randint(0, len(X)-1)] for _ in range(self.k)]

        for _ in range(self.iter):
            cluster_indices = self.assign_cluster(X)
            new_centroids = self.findNewCentroids(X, cluster_indices)
            if new_centroids == self.centroids:
                break
            self.centroids = new_centroids

        # Create list of points per cluster
        clustered_points = [[] for _ in range(self.k)]
        for i, cluster_id in enumerate(cluster_indices):
            clustered_points[cluster_id].append(X[i])

        print("Final centroids:", self.centroids)
        print("Clustered points:", clustered_points)
        return clustered_points

    def assign_cluster(self, X):
        clusters = []
        for point in X:
            distances = [sqrt((point[0]-c[0])**2 + (point[1]-c[1])**2) for c in self.centroids]
            cluster_index = distances.index(min(distances))
            clusters.append(cluster_index)
        return clusters

    def findNewCentroids(self, X, clusters):
        new_centroids = []
        for c in range(self.k):
            assigned_points = [X[i] for i in range(len(X)) if clusters[i] == c]
            if assigned_points:
                dim = len(X[0])
                centroid = [sum(p[d] for p in assigned_points)/len(assigned_points) for d in range(dim)]
                new_centroids.append(centroid)
            else:
                new_centroids.append(self.centroids[c])
        return new_centroids

# Example usage
points = [[1,2],[1,4],[1,0],[10,2],[10,4],[10,0]]
km = Kmeans(k=2, iter=10)
clustered_points = km.fit(points)
```

```python
from math import sqrt
import random
class Kmeans:
    def __init__(self,k=3,iter=5):
        self.k = k
        self.iter = iter
        self.centroids = []
        
    def fit(self, X):

        # for i in range(self.k):
        #     self.centroids.append(X[i])
        for i in range(self.k):
            j = random.randint(0,len(X)-1)
            self.centroids.append(X[j])

        iteration = 1
        new_centroids = []
        while new_centroids!= self.centroids and iteration<=self.iter :
            if(len(new_centroids)!=0):
                self.centroids = new_centroids
                print(f'Centroids{iteration} : ',self.centroids)

            clustures = self.assign_clusture(X)
            #Print Clusture

            print("CLUSTURES : ",'[',sep = '\n')
            for clusture in clustures:
                print(clusture)
            print(']',sep = '\n')

            new_centroids = self.find_new_centroids(clustures,self.k)
            print(f"new centroid:{iteration}",new_centroids)

            iteration = iteration+1


        

    def assign_clusture(self,X):
        matrix = [[None for _ in range(len(self.centroids))] for _ in range(len(X))]
        
        for row in range(len(matrix)):
            L = [None]*len(matrix[0])
            for column in range(len(matrix[0])):
                
                L[column] = sqrt((X[row][0] - self.centroids[column][0])**2 + (X[row][1] - self.centroids[column][1])**2)
            
            L2 = self.keepOnlyMin(L)
            for i in range(len(L2)):
                if L2[i]!=None:
                    # L2[i] = (X[row][0],X[row][1])
                    L2[i] = tuple(X[row])
                
            matrix[row] = L2

        return matrix

    def keepOnlyMin(self, L):
        min_val = min(L)
        new_L = [None if x != min_val else x for x in L]
        return new_L

    def find_new_centroids(self,matrix , k):
        L = [0]*k

        for column in range(len(matrix[0])):
            x_count , y_count = 0.0,0.0
            s = 0.0
            for row in range(len(matrix)):
                if matrix[row][column]!=None:
                    x_count+=matrix[row][column][0]
                    y_count+=matrix[row][column][1]
                    s+=1
            
            L[column] = (x_count/s,y_count/s)

        return L
                
            
points = [
    (1, 1),
    (1.5, 2),
    (3, 4),
    (5, 7),
    (3.5, 5),
    (4.5, 5),
    (3.5, 4.5),
    (8, 7),
    (3,3.3),
    (11,3.8)
]

km = Kmeans()
km.fit(points)

```
