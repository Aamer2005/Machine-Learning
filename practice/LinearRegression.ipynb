```python
import numpy as np
import pandas as pd

df = pd.read_csv(r'C:\Users\hp\OneDrive\Desktop\SKILLS\Machine Learning\practice\employeeData.csv')

print(df)
```

```python
X = df['Hours']
Y = df['Score']

X = np.array(X)
Y = np.array(Y)

print(X)
print(Y)
```

```python
X_mean = X.mean()
Y_mean = Y.mean()

m = np.sum((X-X_mean)*(Y-Y_mean))/np.sum((X-X_mean)**2)

print(m)

c = Y_mean - m*X_mean
print(c)

```

```python
X = df['Hours']
Y = df['Score']
    
X = np.array(X)
Y = np.array(Y)

print(X)
print(Y)
```

```python
iteration = 1000
learning_rate = 0.01
m,c = 0,0

for x in range(iteration):
    error = Y - (m*X+c)
    m0 = -(2/len(Y))*np.sum(X*(error))
    c0 = -(2/len(Y))*np.sum(error)

    m = m-(learning_rate * m0)
    c = c-(learning_rate * c0)

print(m,c)

pred = []
for x in X:
    pred.append(m*x+c)

print(pred)
```
