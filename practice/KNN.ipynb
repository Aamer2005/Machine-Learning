```python
import numpy as np

X = np.array([
    [1, 35],
    [2, 40],
    [3, 50],
    [4, 55],
    [5, 60],
    [6, 65],
    [7, 70],
    [8, 80],
    [9, 85],
    [10, 90]
])

Y = np.array([
    'Fail', 'Fail', 'Fail', 'Fail', 'Pass',
    'Pass', 'Pass', 'Pass', 'Pass', 'Pass'
])

```

```python
newpoint = [10,90]

k = 5
```

```python
X1 = [x[0] for x in X]
print(X1)

X2 = [x[1] for x in X]
print(X2)
```

```python
def cal_distance(x1,x2,px1,px2):
    return np.sqrt((x1-px1)**2 + (x2-px2)**2)
```

```python
distances = []
val = []
count = 0
for x1,x2 in zip(X1,X2):
    distances.append(cal_distance(x1,x2,newpoint[0],newpoint[1]))
    val.append([x1,x2,cal_distance(x1,x2,newpoint[0],newpoint[1])])


print(distances)
print(val)
```

```python
distances.sort()
print(distances)

k_distances = [distance for i,distance in enumerate(distances) if i<k ]

print(k_distances)
k_index = []

for k_d in k_distances:
    for x in val:
        if x[2] == k_d:
            m = X1.index(x[0])
            if m in k_index:continue
            k_index.append(m)
            print(x[2])
            break

print(k_index)
```

```python
count1,count2 = 0,0

for indexi in k_index:
    if Y[indexi] == 'Fail':count1+=1
    if Y[indexi] == 'Pass':count2+=1

result = 'Pass' if count1<count2 else 'Fail' 
print(result)
```
